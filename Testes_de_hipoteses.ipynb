```python
from scipy.stats import norm
import seaborn as sns
def calculo_media(tabela): #calcula a media
  n = len(tabela)
  media = sum(tabela)/n
  return media

def calculo_desvio(tabela): #Calcula o desvio considerando o denominador como n-1
  n = len(tabela)
  media = calculo_media(tabela)
  return (sum([(x - media)**2 for x in tabela])/(n-1))**(1/2)

def calculo_varianca(tabela): #calcula a variancia com tamanho n-1
  n = len(tabela)
  media = calculo_media(tabela)
  return (sum([(x - media)**2 for x in tabela])/(n-1))

def probabilidade_normal(z): #Pega o valor de Z na tabela normal
  return norm.cdf(z)


def aderencia(comparacao,Precisao):
  k = len(comparacao)  # Tamanho da lista 'comparacao'
  print(f"Tamanho de comparacao: {k}")
  print(f"Tamanho de Precisao: {len(Precisao)}")

  # Verificar se os tamanhos de 'Precisao' e 'comparacao' s√£o iguais antes de prosseguir
  if len(Precisao) != k:
      print("Erro: 'Precisao' e 'comparacao' t√™m tamanhos diferentes!")
  else:
      t = 0
      total = 0

      # Loop correto para comparar 'Precisao' e 'comparacao'
      for i, j in zip(Precisao, comparacao):
          if i == j:
              total += 1
          else:
              print(f"Discrep√¢ncia no √≠ndice {t}")
          t += 1

      # Exibe a propor√ß√£o de elementos iguais
      proporcao = (total / k) * 100
      print(f"Propor√ß√£o de elementos iguais: {proporcao:.2f}%")
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats

def test_ks(tabela,significancia):
  tabela.sort()
  n = len(tabela)
  if(n<30):
    print('Tamanho da amostra menor que 30') #verificamos o tamanho da lista para ver se esse √© o teste ideal
    return
  media = calculo_media(tabela)
  desvio_padrao = calculo_desvio(tabela)


  tabela_geral = []
  visitados = []
  for i in tabela:
    if i not in visitados:
      visitados.append(i)


  tabela_geral.append(visitados)


  #F_abs
  F_abs = [tabela.count(i) for i in tabela_geral[0]]
  tabela_geral.append(F_abs)

  #F_ac
  F_ac = []
  for i in F_abs:
    if len(F_ac) == 0:
      F_ac.append(i)
    else:
      F_ac.append(i + F_ac[len(F_ac)-1])
  tabela_geral.append(F_ac)

  #F_obs
  F_obs = [i/n for i in F_ac]
  tabela_geral.append(F_obs)

  #Z_i
  Z_i = [(i-media)/desvio_padrao for i in visitados]
  tabela_geral.append(Z_i)

  #F_esp
  F_esp = [probabilidade_normal(i) for i in Z_i]
  tabela_geral.append(F_esp)

  #F_esp(X_i) - F_obs(X_i)
  F_esp_menos_F_obs = [abs(F_esp[i] - F_obs[i]) for i in range(len(F_esp))]
  tabela_geral.append(F_esp_menos_F_obs)

  #F_esp(X_i) - F_obs(X_(i-1))
  F_esp_menos_F_obs_outro = [abs(F_esp[i]) if i == 0 else abs(F_esp[i] - F_obs[i-1]) for i in range(len(F_esp))]
  tabela_geral.append(F_esp_menos_F_obs_outro)


#Gera√ß√£o de tabela ----------------------------------------------------------------------------------------------------------------------------------------------------------------------------
  colunas = ['Visitados', 'F_abs', 'F_ac', 'F_obs', 'Z_i', 'F_esp', '|F_esp - F_obs|', '|F_esp - F_obs_(i-1)|']
  df = pd.DataFrame({colunas[i]: tabela_geral[i] for i in range(len(colunas))})

  df['Visitados'] = df['Visitados'].map(lambda x: f"{x:.2f}")
  df['F_abs'] = df['F_abs'].astype(int)  # Sem casas decimais
  df['F_ac'] = df['F_ac'].astype(int)  # Sem casas decimais
  df['Z_i'] = df['Z_i'].map(lambda x: f"{x:.2f}")  # 2 casas decimais
  df['F_obs'] = df['F_obs'].map(lambda x: f"{x:.3f}")  # 3 casas decimais
  df['F_esp'] = df['F_esp'].map(lambda x: f"{x:.4f}")  # 4 casas decimais

  # Usar map em vez de applymap
  df['|F_esp - F_obs|'] = df['|F_esp - F_obs|'].map(lambda x: f"{float(x):.7f}" if isinstance(x, (int, float)) else x)
  df['|F_esp - F_obs_(i-1)|'] = df['|F_esp - F_obs_(i-1)|'].map(lambda x: f"{float(x):.7f}" if isinstance(x, (int, float)) else x)
#----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------


  D_cal = max([max(abs(i), abs(t)) for i in F_esp_menos_F_obs for t in F_esp_menos_F_obs_outro])


  if significancia == 0.05:
      De = 1.36/n**(1/2)

  elif significancia == 0.20:
      De = 1.07/n**(1/2)

  elif significancia == 0.15:
      De = 1.14/n**(1/2)

  elif significancia == 0.10:
      De = 1.22/n**(1/2)

  elif significancia == 0.01:
      De = 1.63/n**(1/2)

  else:
    print('Significancia invalida')
    return


  if(D_cal > De):
    # print(f"{D_cal:.3f} > {De:.3f}")
    resultado = 'Hip√≥tese nula rejeitada, ou sejam, n√£o segue uma distribui√ß√£o normal'

  else:
    # print(f"{D_cal:.3f} < {De:.3f}")
    resultado = 'Hip√≥tese nula n√£o rejeitada, ou seja, pode seguir uma distribui√ß√£o normal '


  #resultado
  resultado_df = pd.DataFrame({'Resultado': [resultado]})

  coluna = 'Resposta'
  pd.set_option('display.max_colwidth', None)
  df2 = pd.DataFrame({coluna: [resultado] })
  # display(df2)
  # display(df)





#Vai ser utilizado para o teste de T student
  if(D_cal > De):

    # print("False")
    return False

  else:
    # print("True")
    return True
```

```python
import numpy as np

def gerar_varias_amostras_normais(qtd_amostras=6, tamanho_inicial=31, media=0, desvio_padrao=1, incremento=10):
    amostras = []
    for i in range(qtd_amostras):
        tamanho = tamanho_inicial + i * incremento  # Cada amostra ter√° tamanho maior
        amostra = list(np.random.normal(loc=media, scale=desvio_padrao, size=tamanho))
        amostras.append(amostra)
    return amostras

# Exemplo de uso

import numpy as np

import numpy as np

def gerar_lista_de_amostras_exponenciais(n_amostras, n_pontos, scale=2):
    lista_amostras = []

    for _ in range(n_amostras):
        amostra = np.random.exponential(scale=scale, size=n_pontos)
        lista_amostras.append(amostra.tolist())  # Adicionar a amostra como lista

    return lista_amostras


# Gerar lista de listas de amostras exponenciais
lista_amostras = gerar_lista_de_amostras_exponenciais(n_amostras=100, n_pontos=50)

amostras = gerar_varias_amostras_normais(qtd_amostras=100, tamanho_inicial=31, media=5, desvio_padrao=2, incremento=10)

```

```python
import numpy as np
from scipy import stats

def teste_cochran(*grupos):
    k = len(grupos)  # N√∫mero de grupos
    n = sum(len(grupo) for grupo in grupos)  # Total de observa√ß√µes
    k_mais_um = k - 1

    # Calcular a m√©dia de cada grupo
    medias = [np.mean(grupo) for grupo in grupos]

    # Calcular a vari√¢ncia de cada grupo
    variancias = [np.var(grupo, ddof=1) for grupo in grupos]

    # Calcular a vari√¢ncia total
    variancia_total = sum(variancias)

    # Calcular a estat√≠stica de Cochran
    C = np.max(variancias) / variancia_total

    # Calcular o p-valor usando a distribui√ß√£o F

    p_valor = 1 - stats.f.cdf(C, k_mais_um, n - k)

    return C, p_valor



# Calcular a estat√≠stica de Cochran e o p-valor
estatistica_cochran, p_valor_cochran = teste_cochran(*grupos)

# Imprimindo os resultados
print(f'Estat√≠stica de Cochran: {estatistica_cochran:.4f}, p-valor: {p_valor_cochran:.4f}')

# Interpreta√ß√£o do p-valor
alpha = 0.05
if p_valor_cochran < alpha:
    print("Rejeitamos a hip√≥tese nula: as vari√¢ncias entre os grupos s√£o diferentes.")
else:
    print("N√£o rejeitamos a hip√≥tese nula: as vari√¢ncias entre os grupos podem ser homog√™neas.")
```

##Teste de Levene





$Z_{ij} = |Y_{ij} - \tilde{Y}_i|$

$ Y_{ij}  \text{ √© a j -√©sima observa√ß√£o no grupo i} $

$ {Y}_i \text{ √© a m√©dia das observa√ß√µes no grupo i}$




$W = \frac{(N - k)}{(k - 1)} \cdot \frac{\sum_{i=1}^{k} n_i (\bar{Z}_i - \bar{Z})^2}{\sum_{i=1}^{k} \sum_{j=1}^{n_i} (Z_{ij} - \bar{Z}_i)^2}$



 N  √© o n√∫mero total de observa√ß√µes (somando todos os grupos). \\

 k  √© o n√∫mero de grupos.

 $n_i$  √© o tamanho do grupo  i
  
 $\bar{Z}_i  \text{ √© a m√©dia dos }  Z_{ij} \text{ no grupo }  i $
  
 $\bar{Z}  \text{ √© a m√©dia global dos } Z_{ij}$ .







```python
def test_levene(*tabelas,Significancia):

  N = sum([len(i) for i in tabelas]) # total de observa√ß√µes
  k = len(tabelas)   #grau nde liberdade
  n_i= [len(i) for i in tabelas] #tamanho de cada tabela
  medias = [calculo_media(i) for i in tabelas] #media de cada tabela

  diferencas_absolutas=[[abs(i-medias[t]) for i in tabelas[t]] for t in range(len(tabelas))] #diferenca de cada elemento - sua media em modulo

  soma_do_denominador = []
  for i in diferencas_absolutas:
    soma_do_denominador.extend(i) #vou colocar todos as diferencas em uma so lista, ao inves de uma lista de lista


  medias_novas_z = [calculo_media(i) for i in diferencas_absolutas] #as medias das diferencas

  media_z = calculo_media(medias_novas_z) #media da media das diferencas

  numerador = sum(n_i[i]*(medias_novas_z[i]-media_z)**2 for i in range(len(tabelas)))

  denominador = sum((soma_do_denominador[i] - media_z)**2 for i in range(N))

  W_cal = (N-k)/(k-1)*(numerador/denominador) #estatistica dde levene

  p_value = 1 - stats.f.cdf(W_cal, k-1 , N-k) #usando calculo para pegar o p na distribuicao F
  print(f"W_cal: {W_cal:.4f}")
  print(f"p_value: {p_value:.4f}")
  if p_value < Significancia:
      print("Rejeitamos a hip√≥tese nula: as vari√¢ncias entre os grupos s√£o diferentes.")
  else:
      print("N√£o rejeitamos a hip√≥tese nula: as vari√¢ncias entre os grupos podem ser homog√™neas.")

  print()
```

```python
from scipy import stats
import pingouin as pg
from statsmodels.stats.weightstats import ztest

#PROCURAR O ZTEST
def test_z_caso_bilateral(media_a_ser_comparada, significancia, desvio_da_populacao, tamanho_da_amostra=None, media_da_amostra=None, tabelas=None):
    if tabelas is not None:
        media = calculo_media(tabelas)
        tamanho = len(tabelas)
    elif tamanho_da_amostra is not None and media_da_amostra is not None:
        media = media_da_amostra
        tamanho = tamanho_da_amostra
    else:
        print("Falta de informa√ß√µes para realizar o teste")
        return

    denominador = desvio_da_populacao / (tamanho ** 0.5)
    z = (media - media_a_ser_comparada) / denominador

    valor_critico = stats.norm.ppf(1 - significancia / 2)  # Valor cr√≠tico para o teste bilateral

    if abs(z) > valor_critico:
      # print(f"{abs(z):.4f} > {valor_critico:.4f}")
      # print("Hip√≥tese nula rejeitada, as m√©dias s√£o diferentes.")
      return False
    else:
      # print(f"{abs(z):.4f} <= {valor_critico:.4f}")
      # print("Hip√≥tese nula n√£o rejeitada, as m√©dias podem ser iguais.")
      return True


def teste_z_gabarito(mu_0, sigma, n, media_amostra=None, alpha=0.05):
    # Se media_amostra n√£o for fornecida, gera uma m√©dia igual a mu_0
    if media_amostra is None:
        # Gerar uma amostra aleat√≥ria com m√©dia esperada igual a mu_0
        amostra = np.random.normal(mu_0, sigma, n)
    else:
        # Usar uma amostra com m√©dia fornecida
        amostra = np.random.normal(media_amostra, sigma, n)

    # Realizar o teste Z
    z_stat, p_valor = ztest(amostra, value=mu_0)

    # Verificar se rejeita a hip√≥tese nula com base no valor-p
    if p_valor < alpha:
        # print("Rejeita a hip√≥tese nula (H0).")
        return False
    else:
        # print("N√£o rejeita a hip√≥tese nula (H0).")
        return True
```

###Teste t  Student para uma amostra, quando o desvio-padr√£o populacional ( œÉ) n√£o for conhecido


$T_{\text{calc}} = \frac{\bar{X} - \mu_0}{\frac{S}{\sqrt{n}}} \sim t_{n-1}$

```python

def test_t_desvio_geral(tabelas,media_antiga,sign):

  T_calc = (calculo_media(tabelas) - media_antiga)/(calculo_desvio(tabelas)/len(tabelas)**0.5)

  valor_critico = t_table[len(tabelas)-1][sign]

  if abs(T_calc) < valor_critico:
      return True
  else:
      return False

def gabarito_umaT(amostra,media_populacional):
  t_stat, p_value = stats.ttest_1samp(amostra, media_populacional)
  if p_value < 0.05:
    return False
  else:
    return True
```

###TESTE t DE STUDENT PARA COMPARA√á√ÉO DE DUAS M√âDIAS POPULACIONAIS A PARTIR DE DUAS AMOSTRAS ALEAT√ìRIAS INDEPENDENTES, variancia populacional desconhecida


Caso 1: $\sigma_1^2 \neq \sigma_2^2 $

\\

$
T_{\text{calc}} = \frac{(\bar{X}_1 - \bar{X}_2)}{\sqrt{\frac{S_1^2}{n_1} + \frac{S_2^2}{n_2}}}
$

E o n√∫mero de graus de liberdade √© dado por:

$
\nu = \frac{\left(\frac{S_1^2}{n_1} + \frac{S_2^2}{n_2}\right)^2}{\frac{\left(\frac{S_1^2}{n_1}\right)^2}{n_1 - 1} + \frac{\left(\frac{S_2^2}{n_2}\right)^2}{n_2 - 1}}
$

\\

Caso 2: $ \sigma_1^2 = \sigma_2^2 $

\\
$
T_{\text{calc}} = \frac{(\bar{X}_1 - \bar{X}_2)}{S_p \sqrt{\frac{1}{n_1} + \frac{1}{n_2}}}
$

Onde:

$
S_p = \sqrt{\frac{(n_1 - 1)S_1^2 + (n_2 - 1)S_2^2}{n_1 + n_2 - 2}}
$

$\nu = n_1 + n_2 - 2 $ graus de liberdade.$

```python
def test_t_caso2(*tabelas,alpha):
  # print("As variancias podem ser iguais ")
  s_1 = calculo_varianca(tabelas[0])
  s_2 = calculo_varianca(tabelas[1])
  n_1 = len(tabelas[0])
  n_2 = len(tabelas[1])

  s_p = (((n_1 - 1) * s_1 + (n_2 - 1) * s_2) / (n_1 + n_2 - 2))**(1/2)
  T_calc = (calculo_media(tabelas[0]) - calculo_media(tabelas[1]))/(s_p * (1/n_1 + 1/n_2)**(1/2))
  v = n_1 + n_2 -2 #grau de liberadade


  valor_critico = stats.t.ppf(1 - alpha / 2, v)


  if abs(T_calc) < valor_critico:  # Use valor absoluto
        # print(f"{T_calc:.4f} < {valor_critico}")
        # print('Hip√≥tese nula n√£o rejeitada: as m√©dias podem ser iguais.')
        return True
  else:
        # print(f"{abs(T_calc):.4f} > {valor_critico}")
        # print('Hip√≥tese nula rejeitada: as m√©dias s√£o diferentes.')
        return False
```

```python
def test_t_caso1(*tabelas,alpha):
  # print("S√£o diferentes as variancias ")
  s_1 = calculo_varianca(tabelas[0])
  s_2 = calculo_varianca(tabelas[1])
  n_1 = len(tabelas[0])
  n_2 = len(tabelas[1])

  t_1 = s_1/n_1
  t_2 = s_2/n_2

  Comparacao = s_1/n_1  + s_2/n_2

  T_calc = (calculo_media(tabelas[0]) - calculo_media(tabelas[1]))/(Comparacao)**(1/2)

  v = (Comparacao)**2/((t_1)**2/(n_1 -1) + (t_2)**2/(n_2 -1))

  valor_critico =  stats.t.ppf(1 - alpha / 2, v)

  if abs(T_calc) < valor_critico:  # Use valor absoluto
        # print(f"{T_calc:.4f} < {valor_critico}")
        # print('Hip√≥tese nula n√£o rejeitada: as m√©dias podem ser iguais.')
        return True
  else:
        # print(f"{abs(T_calc):.4f} > {valor_critico}")
        # print('Hip√≥tese nula rejeitada: as m√©dias s√£o diferentes.')
        return False
```

###TESTE t DE STUDENT PARA COMPARA√á√ÉO DE DUAS M√âDIAS POPULACIONAIS A PARTIR DE DUAS AMOSTRAS ALEAT√ìRIAS EMPARELHADAS. Tem que ser normais e terem variancias homogeneas


$\bar{d} = \frac{\sum_{i=1}^{n} d_i}{n}$

$S_d = \sqrt{\frac{\sum_{i=1}^{n} (d_i - \bar{d})^2}{n-1}}$

$T_{calc} = \frac{\bar{d} - \mu_d}{\frac{S_d}{\sqrt{n}}}$

```python
def test_t_emparelhadas(*tabelas, sign):
    # Calcula a diferen√ßa entre as amostras emparelhadas
    d_i = [tabelas[0][i] - tabelas[1][i] for i in range(len(tabelas[0]))]

    # Verifica a normalidade das diferen√ßas
    t1 = det_norm(d_i, sign)
    if not t1:
        # print("N√£o √© necess√°rio continuar pois as diferen√ßas n√£o s√£o normais")
        return

    # Verifica homogeneidade de vari√¢ncias (se necess√°rio para o teste)
    if not Test_bartlett(*tabelas, Significancia=sign):
        # print("O teste exige que as vari√¢ncias sejam possivelmente iguais")
        return

    # Calcula a m√©dia e o desvio padr√£o das diferen√ßas
    d_media = calculo_media(d_i)
    d_desvio = calculo_desvio(d_i)

    # Calcula a estat√≠stica t
    T_calc = d_media / (d_desvio / (len(tabelas[0]) ** 0.5))

    # Graus de liberdade
    v = len(tabelas[0]) - 1


    # Calcula o valor cr√≠tico
    valor_critico = stats.t.ppf(1 - sign / 2, v)

    # Realiza a compara√ß√£o
    if abs(T_calc) < valor_critico:
        # print(f"{T_calc:.4f} < {valor_critico}")
        # print('Hip√≥tese nula n√£o rejeitada: as m√©dias podem ser iguais.')
        return True
    else:
        # print(f"{abs(T_calc):.4f} > {valor_critico}")
        # print('Hip√≥tese nula rejeitada: as m√©dias s√£o diferentes.')
        return False

def gabaritoT_emparelhadas(*amostra,sing):
  t_statistic, p_value = stats.ttest_rel(amostra[0],amostra[1])
  if p_value < sing:
    return False
  else:
    return True
```
